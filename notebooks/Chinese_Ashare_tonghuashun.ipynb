```python
import requests
import time

API_KEY = "sk-fuyao-J_xlNkRFZyK84fGD-2QQWhPYx0pyUura"
BASE_URL = "https://fuyao.aicubes.cn"
```

```python
def get_historical_kline(thscode, start_ms, end_ms, adjust="forward"):
    """
    获取单只标的历史K线
    :param thscode: 如 '159919.SZ'（沪深300ETF）、'600519.SH'
    :param start_ms: 起始时间戳（毫秒）
    :param end_ms: 结束时间戳（毫秒）
    :param adjust: 复权方式，none / forward(前复权) / backward(后复权)
    """
    url = f"{BASE_URL}/api/a-share/prices/historical"
    params = {
        "thscode": thscode,
        "interval": "1d",
        "start": start_ms,
        "end": end_ms,
        "adjust": adjust,
    }
    headers = {"X-api-key": API_KEY}
    
    response = requests.get(url, params=params, headers=headers)
    data = response.json()
    
    if data.get("code") == 0:
        return data["data"]["item"]  # K线列表
    else:
        raise Exception(f"Error {data.get('code')}: {data.get('message')}")
```

```python
# 示例：获取茅台（600519.SH）2025年全年日线数据
start = int(time.mktime((2025, 1, 1, 0, 0, 0, 0, 0, 0)) * 1000)
end = int(time.mktime((2025, 12, 31, 0, 0, 0, 0, 0, 0)) * 1000)
bars = get_historical_kline("600519.SH", start, end)
print(bars[:3])
```

```python
def search_ticker(keyword):
    url = f"{BASE_URL}/api/meta/tickers/search"
    params = {"q": keyword, "limit": 20}
    headers = {"X-api-key": API_KEY}
    response = requests.get(url, params=params, headers=headers)
    return response.json()["data"]["item"]

# 搜索沪深300ETF
results = search_ticker("ETF")
for item in results:
    print(f"{item['thscode']} - {item['name']} ({item['exchange']})")
```
